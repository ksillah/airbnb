```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from  itertools import groupby
```

```python
air_bnb = pd.read_csv('files/AB_NYC_2019.csv')
air_bnb
```

```python
# How many neighborhood groups are available and which shows up the most?

grouped_by_neighborhood= air_bnb.groupby('neighbourhood_group', as_index = False).mean().round(decimals = 2)
print(grouped_by_neighborhood)
count_by_neighborhood=air_bnb.groupby('neighbourhood_group', as_index = False).size()
count_by_neighborhood
#there are 5 neighborhood groups and Manhattan has the most listings
```

```python
# Are private rooms the most popular in manhattan?
n_by_room_type=air_bnb.groupby(['neighbourhood_group', 'room_type']).size()
n_by_room_type
#No. The entire home/apt is most popular
```

```python
# Which hosts are the busiest and based on their reviews?
air_bnb.groupby('host_id', as_index = False).sum()[['host_id','number_of_reviews']].sort_values('number_of_reviews', ascending=False)
#Hosts with the id 37312959, 344035, 26432133
```

```python
#Which neighorhood group has the highest average price?
grouped_by_neighborhood
#manhattan
```

```python
# Which neighbor hood group has the highest total price?
highest_total_price=air_bnb.groupby('neighbourhood_group', as_index = False).sum()
highest_total_price
highest_listing=air_bnb.sort_values('price', ascending=False)
highest_listing
#manhattan listings are the most expensive altogether, but the3  most expensive listings are in Queens, Brooklyn and Manhattan
```

```python
#Which top 5 hosts have the highest total price?
richest_hosts=air_bnb.groupby('host_id', as_index = False).sum()[['host_id','price']].sort_values('price', ascending=False)
richest_hosts
#the 5 richest hosts are the hosts with the following ids 219517861, 107434423, 156158778, 205031545,30283594
highest_listing
#the 5 hosts with the most epensive listings are the hosts with the following ids 20582832, 5143901,72390391,4382127,3906464
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?

no_rentals=air_bnb[air_bnb['availability_365']==0]
no_rentals
good_no_rentals=no_rentals[no_rentals['number_of_reviews']>=100]
good_no_rentals
#162 hosts have no availability and more than 100 reviews
```

```python
# What host has the highest total of prices and where are they located?
richest_hosts
air_bnb[air_bnb['host_id']==219517861]
#Sonder with host id 219517861 has the highest total of prices and they are located in Manhattan
```

```python
# When did Danielle from Queens last receive a review?
danielles=air_bnb[air_bnb['host_name']=='Danielle']
q_danielles=danielles[danielles['neighbourhood_group']=='Queens']
q_danielles
#There are multiple. The one with id 26432133 last received one on 7-8-19. The one with ID 154256662 last received one on 1-2-18
```
